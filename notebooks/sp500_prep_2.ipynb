```python
def bin_tp_label(row, tp):
    window = df[row.name: row.name + pd.Timedelta('24h')]
    returns = (window.close / row.close) - 1
    if returns[returns > tp].shape[0] >= 1:
        return 1
    else:
        return 0
    
tp = 0.004
df[f'tp_{tp}'] = df.apply(lambda row: bin_tp_label(row, tp), axis=1)
df['tp_0.004'].value_counts()
```
